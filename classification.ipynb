```python
#step 2:dataset exploration- fetch and display our dataset in table format
data=pandas.read_csv(r"C:\Users\Sarah\Desktop\modcom\class\kindi-sarah\data science\data\data.csv")
data
```

```python
#step 3: data cleaning - removing null cells
data.isnull().sum()
```

```python
#step 4: data anlysis- we use describe(),groupby() etc to anlyse data
data.describe()
```

```python
#step 5: Classification: step 1: split data
#we are going to split into 2: X and Y, this is bcs we are using supervised data where we train and test models.
array=data.values
array
#x is going to be numerical datatypes
X=array[0:51,0:8]
X
```

```python
#y is for categorical datatypes
Y=array[0:51,8:]
Y
```

```python
#Step2: set X and Y into training data(70%) and testing data(30%)
from sklearn import model_selection
X_train,X_test,Y_train,Y_test=model_selection.train_test_split(X,Y,train_size=0.3,random_state=42) 
```

```python
#step 3: parsing through an algorithm: support Vector Machines
from sklearn.svm import SVC #support vector Classification
model =SVC()
model.fit(X_train,Y_train)
```

```python
#step 4: testing. we test the model to check the accuracy. 
model_prediction=model.predict(X_test)
model_prediction
```

```python
Y_test
```

```python
#here we check the module's accuracy
from sklearn.metrics import accuracy_score
check_accuracy=accuracy_score(model_prediction,Y_test)*100
check_accuracy
```

```python
#this step shows model Cross Validation, where we check the most likely performance when Fitted to our dataset.
# Cross Validation: Determine The best model for a particular dataset
# Imports Classification Algorithms
# Imports Classification Models
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.svm import SVC

# Create an Empty List and append all models, Give the Models an alias Name
# i.e we called KNeighborsClassifier() alias KNC etc
models = []
models.append(('D Trees', DecisionTreeClassifier()))
models.append(('Gaussian', GaussianNB()))
models.append(('KNC', KNeighborsClassifier()))
models.append(('Random Forest', RandomForestClassifier()))
models.append(('Gradient Boosting', GradientBoostingClassifier()))
models.append(('Linear Disc', LinearDiscriminantAnalysis()))
models.append(('Support Machines', SVC(gamma = 'auto')))

# Import Cross Validation and KFOLD
from sklearn.model_selection import cross_val_score, KFold
# Create a for loop so that each model is tested in turn
for name, model in models:
     # Here we do a 10 split K-FOLD
     kfold = KFold(n_splits = 10, random_state=42, shuffle=True)

     # We get the results for each Fold
     cv_results = cross_val_score(model, X_train, Y_train, cv = kfold,
     scoring='accuracy')

     # Get the average of all Folds
     print(name, '  Results:= ', cv_results.mean())
```
