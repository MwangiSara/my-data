```python
#step 2:dataset exploration- fetch and display our dataset in table format
import pandas
advert=pandas.read_csv(r"C:\Users\Sarah\Desktop\modcom\class\kindi-sarah\data science\data\Advertising.csv")
advert
```

```python
#step 5: regression, we test continous datatypes and the ouptut is a continous datatype.
#we first split our dataset into two:X(fetures/predictors), Y(target/output) 
array=advert.values
array
X=array[0:200,1:4]
X
Y=array[0:200,4:5]
Y
```

```python
#take our variables X and Y and split them into training set(70%) and testing set(30%)
from sklearn import model_selection
X_test,X_train,Y_test,Y_train=model_selection.train_test_split(X,Y,test_size=0.3,random_state=42) # test_size specifies the proportion of the dataset to be allocatd to the test and train set. we use float btwn 1.0 and 0.0.
```

```python
#fit our dataseet to our model , we fit X_train and Y_train.
#we will be using linear regression algorithm
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
#testing our model perfomance using X_test data and Compare the prediction with Y_test data
model_prediction=model.predict(X_test)
model_prediction
```

```python
Y_test
```

```python
#lets check the accuracy of our algorithm
from sklearn.metrics import r2_score
print("accuracy=",r2_score(model_prediction,Y_test)*100)
```

```python
# Imports Regression Models
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression


# Kotlin Compiler for terminal
# open-jdk
# VS Code Extensition: coderunner, kotlin


# Create an Empty List and append all models, Give the Models an alias Name
# i.e we called KNeighborsRegressor()) alias KNR etc
models = []
models.append(('D Trees', DecisionTreeRegressor()))
models.append(('KNR', KNeighborsRegressor()))
models.append(('LR', LinearRegression()))
models.append(('Support Machines', SVR(gamma = 'auto')))
models.append(('Random Forest', RandomForestRegressor()))
models.append(('Gradient Boosting', GradientBoostingRegressor()))
models.append(('Support Machines', SVR(gamma = 'auto')))

# Import Cross Validation and KFOLD
from sklearn.model_selection import cross_val_score, KFold
# Create a for loop so that each model is tested in turn
for name, model in models:
          # Here we do a 10 split K-FOLD
          kfold = KFold(n_splits = 10, random_state=42, shuffle=True)

          # We get the results for each Fold, Train and Test using the Folds
          cv_results = cross_val_score(model, X_train, Y_train, cv = kfold,
          scoring='r2')

          # Get the average of all Folds for current model in the loop
          print(name, ' Results:= ', cv_results.mean())
```
