```python
#first we have to clean our dataset. 
#we have to check if the dataset has any empty cells(rows or columns)
bank.isnull()
#get the sum of total null cells
bank.isnull().sum()
```

```python
#adding unknown to the null cells
bank["Self_Employed"].fillna('unknown',inplace=True) #fillna() fills the empty cells with the unknown
bank.isnull().sum() #isnull() checks the empty/null cells in the dataset
```

```python
#quantitative data(continous) we can get the sum, mean, mode, average, and median with columns
mean_loan=bank["LoanAmount"].mean()
mean_loan
bank["LoanAmount"].fillna(mean_loan,inplace=True)
bank.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x=bank["Gender"],palette='Reds') #we are creating a bar graph for the column gender
plt.title("Distribution of gender") # we are giving the graph a title
plt.ylabel("Number of Members") # we are labeling the y axis

```

```python
#creating countplot() for column Married
sns.countplot(x=bank["Married"]) #countplot() is a univariant plot, just like pie chart
plt.title("Distibution of married members")
plt.ylabel("Number of members")
plt.xlabel("Members")
```

```python
#smarmplot() is a type of scatter plot that is used for representing categorical datatypes. it is bivariante(2 columns/variables) in nature, but it can be multivariante(3 or more columns/variable) when we add a parameter 'hue'
sns.swarmplot(x='Married',y='ApplicantIncome',data=bank,palette='magma',hue='Gender')
```

```python
#we are geting a swamp visualization
sns.swarmplot(x='Gender',y="ApplicantIncome",data=bank, palette = 'magma', hue= 'Self_Employed')
```

```python
#barplot() shows the relationship btwn quantitative and qualitative columns/datatypes. it is bivariante and can also be multivariante if we add the parameter 'hue'
sns.barplot(x='Gender',y='LoanAmount',data=bank,palette='magma',hue='Property_Area')
```

```python
sns.barplot(x='Married',y='CoapplicantIncome',data=bank,hue='Education',palette='magma')
```
